```python
engine = create_engine("mysql+pymysql://" + default_args['db_username'] + ":" + default_args['db_password'] + "@" + default_args['db_hostname'] + "/" + default_args['db_database'] + "", echo=False)
conn = engine.connect()
```

```python
# db = dbUtil(default_args['db_hostname'], default_args['db_username'], default_args['db_password'], default_args['db_database'], True)
sql_string = """
select 
a.business_number, 
a.business_title,
a.address_number,
a.address_name,
a.zipcode,
-- a.boro_id,
c.boro_name,
a.record_date,
b.grade, 
b.grade_date
from dim_business a 
join dim_grade b on a.business_number = b.business_number
join dim_boro c on a.boro_id = c.boro_id
where b.grade_date <= '2016-03-01'
"""
df = pd.read_sql(sql_string, conn)
```

```python
grade_dict = {'A': 5, 'B': 4, 'C': 3, 'P': 2, 'Z': 1}
df['grade_number'] = df['grade'].replace(grade_dict)
```

```python
#df.head(10)
df_grade = df.loc[df['grade'] != 'Not Yet Graded', ['business_number', 'boro_name', 'zipcode', 'grade_number', 'grade_date']]
```

```python
df_grade['grade_number'] = df_grade['grade_number'].astype('int32')
#df_grade[df_grade['grade_number'].isna() == False].count()
#df_grade[df_grade[['grade_number']].applymap(np.isreal) == True].info()
```

```python
zip_mean = df_grade.groupby(['business_number', 'boro_name', 'zipcode'], as_index=False).agg({'grade_number': ['mean', 'min', 'max', 'size']})
zip_mean.columns = ['business_number', 'boro_name', 'zipcode', 'grade_avg', 'grade_min', 'grade_max', 'grade_count']
# zip_mean
```

```python
zip_mean.sort_values(by=['zipcode', 'grade_avg', 'grade_count'], ascending=[True, False, False])
# zip_mean.sort_values(by=['grade_count', 'grade_avg'], ascending=[False, False])
```
